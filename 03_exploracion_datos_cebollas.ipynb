```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# comprobar qué es plt (opcional)
print("plt ->", plt)

#cargar dataset de partidos de futbol
df_partidos = pd.read_csv("partidos_cebollitas_fe82a1a4-e109-41b1-8b78-d9b4341dacaf.csv")


# Inspeccionar datos
df_partidos.head(6)
```

```python
#Estadisticas descriptivas generales
df_partidos.describe()
```

```python
# Calculando promedios de goles usando los nombres correctos de las columnas
# goles_local_cebo = df_partidos[df_partidos['equipo_local_Cebollitas FC'] == 'goles_local'][goles_local].mean()
# goles_visitante_cebo = df_partidos[df_partidos['equipo_visitante_Cebollitas FC'] == 'Cebollitas FC']['goles_visitante'].mean()
goles_local_cebo = df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goles_visitante_cebo = df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print(f"Goles promedio de Cebollitas FC como local: {goles_local_cebo}")
print(f"Goles promedio de Cebollitas FC como visitante: {goles_visitante_cebo}")
```

```python
#Histograma de goles anotados por Cebollitas FC (Local y visitante)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC']['goles_local'], bins=range(0, 10), kde=False, ax=ax[0])
ax[0].set_title('Goles anotados por Cebollitas FC como Local')
ax[0].set_ylabel('Frecuencia')
ax[0].set_xlabel('Goles')

sns.histplot(df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'], bins=range(0, 10), kde=False, ax=ax[1])
ax[1].set_title('Goles anotados por Cebollitas FC como Visitante')
ax[1].set_ylabel('Frecuencia')
ax[1].set_xlabel('Goles')

plt.tight_layout()
plt.show()
```

```python
#Boxplot de goles anotados por Cebollitas FC (Local y visitante)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

sns.boxplot(x='equipo_local', y='goles_local', data=df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC'], ax=ax[0])
ax[0].set_title('Boxplot de Goles Anotados por Cebollitas FC como Local')
ax[0].set_ylabel('Goles')

sns.boxplot(x='equipo_visitante', y='goles_visitante', data=df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC'], ax=ax[1])
ax[1].set_title('Boxplot de Goles Anotados por Cebollitas FC como Visitante')
ax[1].set_ylabel('Goles')

plt.tight_layout()
plt.show()
```

```python
# Crear el scatterplot de posesion_local vs goles_local para Cebollitas FC
sns.scatterplot(x='posesion_local (%)', y='goles_local', data=df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC'])
plt.title('Relación entre Posición y Goles Anotados por Cebollitas FC como Local')
plt.xlabel('Posición Local (%)')
plt.ylabel('Goles Anotados')
plt.show()
```

```python
#Crear scatterplot de posision_visitante vs goles_visitante para Cebollitas FC
sns.scatterplot(x='posesion_visitante (%)', y='goles_visitante', data=df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC'])
plt.title('Relación entre Posición y Goles Anotados por Cebollitas FC como Visitante')
plt.xlabel('Posición Visitante (%)')
plt.ylabel('Goles Anotados')
plt.show()
```

```python
#Mapa de calor de correlacion entre variables clave para Cebollitas FC
plt.figure(figsize=(10, 6))
sns.heatmap(df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC'][['goles_local', 'goles_visitante', 'posesion_local (%)', 'posesion_visitante (%)', 'tiros_arco_local', 'tiros_arco_visitante']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación entre Variables Clave para Cebollitas FC como Local')
plt.show()
```
